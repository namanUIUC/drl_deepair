```python
def mangle(s):
    return s.strip()[1:-1]
```

```python
def cat_json(output_filename, input_filenames):
    with open(output_filename, "w") as outfile:
        first = True
        for infile_name in input_filenames:
            with open(infile_name) as infile:
                if first:
                    outfile.write('[')
                    first = False
                else:
                    outfile.write(',')
                outfile.write(mangle(infile.read()))
        outfile.write(']')
```

```python
cat_json('new.txt', ['JSON_ep1/AvailabilityRequest.txt','JSON_ep1/AvailabilityResponse.txt'])
```
